```python
import pandas as pd
import numpy as np
data = pd.read_csv('pcadata.csv')
```

```python
data.head()
```

```python
# x = all columns - end column
x = data.drop("outcome",axis=1)
# y = end column
y = data['outcome']
# كدا قسمت الداتا ل x , y ( feature , label)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test=train_test_split(x,y,test_size=0.3 ,random_state=0)  
# لازم بالترتيب                                         feature label  النسبة الي هقسم بيها(30و70)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```python
explained_variance = pca.explained_variance_ratio_
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=2, random_state=0)
classifier.fit(x_train, y_train)     # كدا هخليه يتعلم 

y_pred = classifier.predict(x_test)# هسجل الاجابة 
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred )
print(cm)
print(acc)
```
